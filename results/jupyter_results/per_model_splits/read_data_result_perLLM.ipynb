```python
import pandas as pd
import numpy as np

csv_path = 'rf_light_throughput_cv_summary.csv'

df = pd.read_csv(csv_path)
df
```

```python
df_formatted = df.copy()

df_formatted["n_samples"] = df_formatted["n_samples"].map("{:,}".format)

for col in [
    "mean_R2", "std_R2",
    "mean_MAE", "std_MAE",
    "mean_RMSE", "std_RMSE",
    "mean_MRE(%)", "std_MRE(%)"
]:
    df_formatted[col] = df_formatted[col].astype(float).round(3)

df_formatted
```
